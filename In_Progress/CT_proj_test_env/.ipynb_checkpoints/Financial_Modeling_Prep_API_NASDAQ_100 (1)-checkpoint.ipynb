```python
ticker_colum = response_df['symbol']
ticker_colum
```

```python
ticker_names = []

for name in ticker_colum:
  ticker_names.append(name)
# ticker_names
```

```python
def chunks(l, n):
    # For item i in a range that is a length of l,
    for i in range(0, len(l), n):
        # Create an index range for l of n items:
        yield l[i:i+n]


five_ticks = list(chunks(ticker_names, 5))
five_ticks
```

```python
def converttostr(input_seq, seperator):
   # Join all the strings in list
   final_str = seperator.join(input_seq)
   return final_str

seperator = ','
tickers = []

for ticker in five_ticks:
  tickers.append(converttostr(ticker, seperator))

tickers
```

```python
api_key = "96b47c522f00a29fe1b11cfd2b0b02d9"

urls = []

for row in tickers:
  url_history = "https://financialmodelingprep.com/api/v3/historical-price-full/" + row + "?apikey="
  urls.append(url_history)
urls
```

```python
url_0 = requests.request("GET", urls[0]).json()
url_0

# print(urls[0] + api_key)
```
